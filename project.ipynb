```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def create_dataset():
    np.random.seed(42)
    products = ["Laptop", "Phone", "Headphones", "Keyboard", "Mouse"]
    months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
    data = {
        "Product": np.random.choice(products, 50),
        "Month": np.random.choice(months, 50),
        "Sales": np.random.randint(5000, 20000, 50)
    }
    return pd.DataFrame(data)
```

```python
def analyze_sales(df):
    product_sales = df.groupby("Product")["Sales"].sum()
    monthly_sales = df.groupby("Month")["Sales"].sum()
    best_product = product_sales.idxmax()
    best_month = monthly_sales.idxmax()
    return product_sales, monthly_sales, best_product, best_month
```

```python
def plot_sales(product_sales, monthly_sales, months):
    plt.figure(figsize=(10, 4))
    plt.subplot(1, 2, 1)
    product_sales.plot(kind="bar", color="skyblue")
    plt.title("Total Sales per Product")
    plt.ylabel("Sales")
    plt.subplot(1, 2, 2)
    monthly_sales = monthly_sales.reindex(months)
    monthly_sales.plot(kind="line", marker="o", color="green")
    plt.title("Monthly Sales Trend")
    plt.ylabel("Sales")
    plt.tight_layout()
    plt.show()
```

```python
df = create_dataset()
product_sales, monthly_sales, best_product, best_month = analyze_sales(df)

print("Sample Dataset:\n", df.head())
print("\nTotal Sales per Product:\n", product_sales)
print("\nTotal Sales per Month:\n", monthly_sales)
print(f"\nBest Selling Product: {best_product}")
print(f"Best Month for Sales: {best_month}")

plot_sales(product_sales, monthly_sales, ["Jan", "Feb", "Mar", "Apr", "May", "Jun"])
```
